```python
#converting column names into lowercase
df.columns = df.columns.str.lower()
```

```python
#creating a value counts dictionary
value_counts_dict = {col: df[col].value_counts() for col in df.select_dtypes(include='object').columns}

#Printing value counts for each column
for col, value_counts in value_counts_dict.items():
  print("Value counts for column", col, ":")
  print(value_counts)
  print("*" * 50)
```

```python
#replacing unknown/other in race with other

df['race'] = df['race'].replace('Other/Unknown', 'Other')
```

```python
df['race'].value_counts()
```

```python
#replacing unknown region with the mode

region_mode = df['region'].mode()[0]
df['region'] = df['region'].replace('Other/Unknown', region_mode)
```

```python
df['region'].value_counts()
```

```python
#replacing the unknown ethnicity also with the mode

ethnicity_mode = df['ethnicity'].mode()[0]
df['ethnicity'] = df['ethnicity'].replace('Unknown', ethnicity_mode)
```

```python
df['ethnicity'].value_counts()
```

An entry in ntm_speciality says "OBSTETRICS & OBSTETRICS & GYNECOLOGY & OBSTETRICS & GYNECOLOGY"

This seems to be an error, hence it will be replaced with "OBSTETRICS AND GYNECOLOGY"

```python
df['ntm_speciality'] = df['ntm_speciality'].replace('OBSTETRICS & OBSTETRICS & GYNECOLOGY & OBSTETRICS & GYNECOLOGY', 'OBSTETRICS AND GYNECOLOGY')
```

```python
df[df['ntm_speciality'] == "OBSTETRICS & OBSTETRICS & GYNECOLOGY & OBSTETRICS & GYNECOLOGY"]
```

```python
#removing the columns with >40% unknown entries
cols_to_drop = ["risk_segment_during_rx",
                "change_risk_segment",
                "tscore_bucket_during_rx",
                "change_t_score"]

df = df.drop(cols_to_drop, axis=1)
```
